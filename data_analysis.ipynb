```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
df = data.copy()
df['time'] = pd.to_datetime(df['time'])

# set min value to seconds
df['time'] = df['time'].dt.floor('s')
```

```python
df
```

```python
plt.scatter(x = df['time'], y = df['temp'])
plt.plot(df['time'], df['temp'])
plt.title('Temperature over time')
plt.xlabel('Time')
plt.ylabel('Temperature in Â°C')
plt.show()
```

```python
plt.scatter(x = df['time'], y = df['hum'])
plt.plot(df['time'], df['hum'])
plt.title('Relative humidity over time')
plt.xlabel('Time')
plt.ylabel('Relative humidity in %')
plt.show()
```

```python
plt.scatter(x = df['time'], y = df['dist'])
plt.plot(df['time'], df['dist'])
plt.title('Distance over time')
plt.xlabel('Time')
plt.ylabel('Distance in cm')
plt.show()
```
